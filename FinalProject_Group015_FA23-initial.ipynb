# Research Question

Do students at UCSD expect higher grades than they receive if they recommend the professor, as measured through CAPEs? Namely, do students that recommend a professor generally expect to do better (expected grade) than they actually did (grade received) and vice versa?



## Background and Prior Work

CAPES stands for “Course And Professor Evaluations” and are offered to students at the end of every quarter to give feedback on their courses and the professors that taught them. As stated on the website, CAPES<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) is: 

    “a student-run organization that administers a standardized evaluation of UCSD's undergraduate courses and professors. Student feedback gauges the caliber of both the University's curriculum and its faculty. We provide students with the opinions of their peers on any particular course or professor.” (UCSD)
    
After students submit their CAPES, this data is aggregated for each quarter of evaluations into features like the percentage of students that recommended the class, the percentage of students that recommend the instructor, the number of hours per week involved in this class, the average grade expected, and the average grade received (which is calculated from the final submitted grades). CAPEs play an important role in helping students choose their professors and even the courses they enroll them in as they generally have more statistics and comprehensive data compared to other rating sites. 

In addition, other rating sites often show signs of significant skew, with a majority of ratings being either towards the extreme since people are more likely to rate something they have a strong reaction towards. This creates a bimodal distribution that lacks data in between these two strong opinions. On the other hand, while not entirely protected, CAPEs likely have some resistance to potential skews in data since students are often incentivized to complete them for extra credit, which gives equal opportunity for various kinds of responses.

```python
#Function to make the Average Grade Expected and Average Grade Recieved into floats of just the grade point
#number
def grade_standardize(grade):
    
    #ensure grade has not been standardized
    if type(grade) == float:
        return grade
    
    #retain only part after open parenthesis
    grade = grade.split("(")[-1]
    
    #remove close parenthesis
    grade = grade.replace(")", "")
    
    #change to float
    grade = float(grade)
    
    return grade
    
```

```python
capes["Average Grade Expected"] = capes["Average Grade Expected"].apply(grade_standardize)
capes["Average Grade Received"] = capes["Average Grade Received"].apply(grade_standardize)
```

```python
#function to remove % symbol from Percentage Recommended Class and Percentage Recommended Professor and make
#into float

def percent_standardize(percent):
    #ensure percent has not been standardized
    if type(percent) == float:
        return percent
    
    #remove % sign
    percent = percent.replace("%", "")
    
    #convert to float
    percent = float(percent)
    
    return percent
        
```

```python
capes["Percentage Recommended Class"] = capes["Percentage Recommended Class"].apply(percent_standardize)
capes["Percentage Recommended Professor"] = capes["Percentage Recommended Professor"].apply(percent_standardize)
```

```python
capes.dtypes
```

```python
#create new column called difference which is the difference between the average
#grade expected and the average grade recieved
#negative numbers in difference signify students overestimated their grade, positive number signifies students
#underestimated their grade
capes = capes.assign(Difference = capes["Average Grade Received"] - capes["Average Grade Expected"])
```

```python
#create columns for year and terms where each year will be an int (eg. 07, 08, 23, etc) and terms will be
#denoted by numbers(see code)
#this code is taken from https://github.com/COGS108/FinalProjects-Wi21/blob/main/FinalProject_group084.ipynb
```

```python
def extract_term(term):
    TERMS = { "WI": 1, "SP": 2, "S1": 3, "S2": 4, "S3": 5, "FA": 6, "SU" : 7}
    
    term = term[:2]
    term = TERMS[term]
    
    return term
```

```python
def extract_year(qtr):
    
    year = qtr[2:]
    year = int("20" + year)
    
    return year


```

```python
capes["Term"] = capes["Quarter"].apply(extract_term)
capes["Year"] = capes["Quarter"].apply(extract_year)
```

## Linear Model 

Now that we have an understanding of the data we are working with, we will run a regression to understand the correlation between the difference between the average expected and recieved grades and professor recomendation percentage.

However, before we delve into the linear model, we will test for the differnce in means of average expected grades and average received grades to compare the two distributions.

In our intital analysis we found that the average difference between the recieved and expected grade is -0.207682 which indicates that the average grade recieved by the class is 0.21 points lower than averagely expected by the class. 

### Comparing distributions using t-test

```python
expected_grade = capes["Average Grade Expected"]
recieved_grade = capes["Average Grade Received"]

avg_expected = expected_grade.mean()
avg_recieved = recieved_grade.mean()

print("The average grade expected is " + str(avg_expected))
print("The average grade recieved is " + str(avg_recieved))
```

```python
t_val, p_val = ttest_ind(expected_grade,recieved_grade)

print(p_val)
```

Since the p-value calculated above is less than 0.01, we can assume that the difference in the means of the avergae grade expected and the average grade received is statistically significant.

### Linear regression

Now we will run a linear model regression to see the affect professor recomendation has on the average grade expected, average grade recieved and their difference.

```python
#renaming columns for use in patsy.dmatrices

capes = capes.rename(columns = {"Average Grade Expected" : "expectedGrade" ,
                                "Percentage Recommended Professor": "profRcmnd",
                                "Average Grade Received" : "recievedGrade" })
```

```python
# running a linear regression on the effect of professor recomendation and Average grade expected

outcome_1, predictors_1 = patsy.dmatrices("expectedGrade ~ profRcmnd", capes)
mod_1 = sm.OLS(outcome_1,predictors_1)
res_1 = mod_1.fit()

print(res_1.summary())
```

```python
# running a linear regression on the effect of professor recomendation and Average grade recieved 
outcome_2, predictors_2 = patsy.dmatrices("recievedGrade ~ profRcmnd", capes)
mod_2 = sm.OLS(outcome_2,predictors_2)
res_2 = mod_2.fit()

print(res_2.summary())
```

```python
# running a linear regression on the effect of professor recomendation on the difference in 
# average expected and recieved grade

outcome_3, predictors_3 = patsy.dmatrices("Difference ~ profRcmnd", capes)
mod_3 = sm.OLS(outcome_3,predictors_3)
res_3 = mod_3.fit()

print(res_3.summary())
```

The three regressions run above all yield statistically significant results for the effect that recommending a professor has on average recieved grades, average expected grades, and their difference.


The results of the first analysis reveals that for every percentage point a professor is recommended, the average grade expected in the class goes up by 0.0079. 

The results of the second analysis reveals that for every percentage point a professor is recommended, the average grade received in the class goes up by 0.0082.

The results of the final analysis reveals that for every percentage point a professor is recommended, the difference in average grade recieved and the average grade expected in the class goes up by 0.0003. 

Hence, it is reasonable to conclude that "liking" a professor or recommending them has a larger impact on the average grade recieved as compared to the average grade expected however that difference is not large and is seen in the coefficient on profRcmnd in the regression of difference on professor recommendation.  


Digging deeper, we check if being a stem class has any bearing on expected and recieved grades in additon to the professor recommendation percentage as stem classes are percieved to be harder and hence are beleived to have, on average, lower grades.

```python
# creating a column called stem where classes that are classified stem will have value = 1, 
# and rest will have value = 0

capes["stem"] = 0
capes.loc[capes['Category'] == 'stem', 'stem'] = 1

capes["stem"].describe()
```

```python
outcome_4, predictors_4 = patsy.dmatrices("expectedGrade ~ profRcmnd + stem", capes)
mod_4 = sm.OLS(outcome_4,predictors_4)
res_4= mod_4.fit()

print(res_4.summary())
```

```python
outcome_5, predictors_5 = patsy.dmatrices("recievedGrade ~ profRcmnd + stem", capes)
mod_5 = sm.OLS(outcome_5,predictors_5)
res_5= mod_5.fit()

print(res_5.summary())
```

The analyses above also all yield statistically significant results.

We can see that by accounting for a class being stem, the effect of professor recommendation on expected grade decreases from 0.0079 to 0.0067. Being a stem class reduces the expected grade further by -0.1656.

Similarly, the effect of professor recommendation on recieved grade decreases from 0.0082 to 0.0067. Being a stem class reduces the expected grade further by -0.2193.

This tells us that being a stem class has a larger effect on expected and recieved grades than professor recommendation. However, this does not completely diminish the effect of professor recommendation as the results are still statistically significant and show an impact nonetheless.

# Ethics & Privacy

Our research project is focused on gathering student responses to questions about professors and the classes they taught. Since students have the option to answer the evaluations and it is anonymous, there is no risk of not having consent or being affected since they have anonymity. However, there is a risk when it comes to the professors since their full name and the class they taught is displayed. For the sake of our experiment, we chose to not include their names out of respect for their privacy. Furthermore, CAPEs data is publically available. 

Since it is a voluntary survey and the fact that we believe most responses are those with more positive or negative feelings towards the professor/ class, we do believe the results may be slightly biased. We hope to counter this with the fact that the data set includes every class for every quarter for over 15 years. We hope the amount of information we have obtained will slightly offset some of the bias that may appear. However, we acknowledge it is impossible to find or completely eliminate any bias. In conclusion, we are confident there is enough information present to effectively answer our question and we aim to protect professors’ privacy by not including their names.


# Discussion and Conclusion

Our project aims to explore the relationships between student grades (expected and received) and their recommendation of their professor. We hypothesized that students in classes of well-liked professors will overestimate their grade and those in classes of professors who are not as well-liked will underestimate their grade. We used CAPES data from UCSD which include important variables necessary for the exploration of our hypothesis. The data includes the percentage of students that recommended the class, the percentage of students that recommend the instructor, the number of hours per week involved in this class, the average grade expected, and the average grade received (which is calculated from the final submitted grades). 

We used exploratory data analysis to find any trends and relationship between the given variables. By looking at histograms of the percentage that recommend the class and recommend the professor, we found that they had roughly the same left skewed distribution, with most of the recommendations being in the 80-100 range for both variables. Looking at another histogram of the average grade expected and the average grade received, we found a similar distribution with a left skew and most students expecting and receiving around a 3.5 grade point in the class. To investigate any linear relationships between percentage recommended vs. grade received as well as average grade received vs. expected, we created scatterplots to analyze any correlation. We found that the regression line for the percentage recommended vs. grade received was horizontal, suggesting a lack of correlation. The regression line for average grade received vs. expected was more positive, suggesting a positive correlation and that students were accurately predicting their grades. A scatterplot was also made for difference in grades expected and received vs percentage of recommendations of the professor. We found that a majority of the plots clustered around grade difference 0.0 to 0.5, suggesting that people perhaps tend to accurately estimate their performance. The plots also occur around high recommendations of the professor, being 80-100. But there is no clear negative or positive correlation in the relationship.
